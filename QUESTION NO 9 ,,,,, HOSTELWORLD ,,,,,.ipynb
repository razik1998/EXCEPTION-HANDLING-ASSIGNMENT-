```python
#  IMPORTING REQUIRED LABRARIES

import selenium 
import pandas as pd 
from selenium import webdriver
import warnings
warnings.filterwarnings ('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException, TimeoutException
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import requests
import time
```

```python
#  CONNECTING TO THE DRIVER
driver=webdriver.Chrome(r"C:\Users\ASUS\Downloads\chromedriver.exe")
```

```python
#  OPENING THE HOSTELWORLD PAGE ON AUTOMATED CHROME BROWSER
driver.get("https://www.hostelworld.com/")
```

```python
#   ENTERING PRODUCT AS REQUIRED 
products = driver.find_element (By.XPATH,"/html/body/div[3]/div/div/div[2]/div[2]/div/div/div[4]/div/div[2]/div/div[1]/div/div/div/input ")
products.send_keys ('LONDON')
```

```python
#  CLICKING SEARCH BUTTON 
search = driver.find_element (By.XPATH," /html/body/div[3]/div/div/div[2]/div[2]/div/div/div[4]/div/div[2]/div/div[1]/div/div/ul/li[2]/div")
search.click()
```

```python
#  CLICKING SEARCH BUTTON 
search = driver.find_element (By.XPATH," /html/body/div[3]/div/div/div[2]/div[2]/div/div/div[4]/div/div[2]/div/div[5]")
search.click()
```

```python
#  SCRAPING ALL HOSTEL NAME
HOSTEL_NAME = []
try:
    hos = driver.find_elements (By.XPATH,'//h2[@class="title title-6"]')
    for i in hos:
        temp = i.text
        HOSTEL_NAME.append (temp)
except NoSuchElementException:
        HOSTEL_NAME.append ('-')
        
```

```python
#  PRINTING LENGHT OF ELEMENTS 
print(len(HOSTEL_NAME))
```

```python
#  SCRAPING ALL DISTANCE FROM CITY CENTER 
DIS_CITY = []
try:
    dfc = driver.find_elements (By.XPATH,'//span[@class="description"]')
    for i in dfc:
        temp = i.text
        DIS_CITY.append (temp)
except NoSuchElementException:
        DIS_CITY.append ('-')
        
```

```python
#  PRINTING LENGHT OF ELEMENTS 
print(len(DIS_CITY))
```

```python
#  SCRAPING ALL RATINGS
RATINGS = []
try:
    rtg = driver.find_elements (By.XPATH,'//div[@class="score orange big"]')
    for i in rtg:
        temp = i.text
        RATINGS.append (temp)
except NoSuchElementException:
        RATINGS.append ('-')
        
```

```python
#  PRINTING LENGHT OF ELEMENTS 
print(len(RATINGS))
```

```python
#  SCRAPING ALL TOTAL REVIEWS
REVIEWS = []
try:
    rws = driver.find_elements (By.XPATH,'//div[@class="reviews"]')
    for i in rws:
        temp = i.text
        REVIEWS.append (temp)
except NoSuchElementException:
        REVIEWS.append ('-')
        
```

```python
#  PRINTING LENGHT OF ELEMENTS 
print(len(REVIEWS))
```

```python
#  SCRAPING ALL OVERALL REVIEWS
OVERALL_REVIEWS = []
try:
    orws = driver.find_elements (By.XPATH,'//div[@class="keyword"]')
    for i in orws[0:30]:
        temp = (i.text)
        OVERALL_REVIEWS.append (temp)
except NoSuchElementException:
        OVERALL_REVIEWS.append ('-')
        
```

```python
#  PRINTING LENGHT OF ELEMENTS 
print(len(OVERALL_REVIEWS))
```

```python
#  SCRAPING ALL PRIVATES FROM PRICE
P_PRICE = []
try:
    ppr = driver.find_elements (By.XPATH,'//div[@class="price title-5"]')
    for i in ppr[0:30]:
        temp = (i.text)
        P_PRICE.append (temp)
except NoSuchElementException:
        P_PRICE.append ('-')
        
```

```python
#  PRINTING LENGHT OF ELEMENTS 
print(len(P_PRICE))
```

```python
#  SCRAPING ALL DORMS FROM PRICE
D_PRICE = []
try:
    dpr = driver.find_elements (By.XPATH,'//div[@class="price title-5"]')
    for i in dpr[0:30]:
        temp = (i.text)
        D_PRICE.append (temp)
except NoSuchElementException:
        D_PRICE.append ('-')
        
```

```python
#  PRINTING LENGHT OF ELEMENTS 
print(len(D_PRICE))
```

```python
#  SCRAPING ALL FACILITIES 
FACILITIES  = []
try:
    fct = driver.find_elements (By.XPATH,'//span[@class="label"]')
    for i in fct[0:30]:
        temp = (i.text)
        FACILITIES .append (temp)
except NoSuchElementException:
        FACILITIES .append ('-')
        
```

```python
#  PRINTING LENGHT OF ELEMENTS 
print(len(FACILITIES))
```

```python
#  SCRAPING ALL PROPERTY DESCRIPTION  
P_DESCRIPTION   = []
try:
    pdc = driver.find_elements (By.XPATH,'//span[@class="label"]')
    for i in pdc[0:30]:
        temp = (i.text)
        P_DESCRIPTION  .append (temp)
except NoSuchElementException:
        P_DESCRIPTION  .append ('-')
        
```

```python
#  PRINTING LENGHT OF ELEMENTS 
print(len(P_DESCRIPTION ))
```
